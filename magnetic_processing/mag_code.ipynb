```python
import pandas as pd
import matplotlib.pyplot as plt
from pyproj import Transformer, CRS
import numpy as np
from math import atan2,pi,dist,sin,cos,sqrt,log
from os import listdir
from pathlib import Path
from sys import stdout
from shapely.geometry import Point,Polygon,LineString
from statsmodels.stats.stattools import durbin_watson
import os
os.environ['USE_PYGEOS'] = '0'
from geopandas import GeoSeries,GeoDataFrame
```

Instantiating a MagApp class which is the interface to the library code which performs the data manipulation.

MagApp Attributes:

    * parameters: python dictionary of input variable
    * data: A pandas dataframe created from the "file" parameter
    * lines: a dictionary of pandas dataframes where each key:value pair are a single line separeate from data,which starts out unpopulated

```python
class MagApp:
    def __init__(self, parameters,merged=None):

        if parameters != None and merged is None:
            self.parameters = parameters
        
            self.data = pd.read_csv(parameters['filepath'],sep=' ')

            self.lines = None

        else:
            self.parameters = parameters
            self.data = merged
            self.lines = None
            
    def transform_coords(self):
        in_crs = CRS.from_epsg(self.parameters['input_epsg'])
        out_crs = CRS.from_epsg(self.parameters['output_epsg'])

        transformer = Transformer.from_crs(in_crs, out_crs)
        self.data["Easting"], self.data["Northing"] = transformer.transform(self.data.Lat.values, self.data.Long.values)
    
    def get_distance(self) -> None:

            distances = []
            length = len(self.data.index)
            cols = self.data.columns
            self.data.sort_values(by=['Northing'], ascending=True).reset_index(drop=True)

            if 'Easting' in cols:
                for i in range(length-1):
                    pointa = (
                        self.data.Easting.values[i], self.data.Northing.values[i])
                    pointb = (
                        self.data.Easting.values[i+1], self.data.Northing.values[i+1])

                    distances.append(dist(pointa, pointb))
            else:
                print('no easting')
                exit()

            distances.insert(0, 0)

            self.data["Dist"] = distances
            
    def split_lines(self, cutoff_length=99, cutoff_dist=None):
        
            if cutoff_dist == None:
                cutoff_dist = self.data.Dist.mean()+5

            line_dict = {}
            t = 0
            k = 1
            length = len(self.data.index)
            i = 0
            while i <= length-1:
                if self.data.Dist.values[i] > cutoff_dist:
                    if k > 9:
                        key = 'line_'+str(k)
                    else:
                        key = 'line '+str(k)
                    line_dict[key] = self.data.iloc[t:i]
                    t = i
                    k += 1
                if i == length-1:
                    if k > 9:
                        key = 'line_'+str(k)
                    else:
                        key = 'line '+str(k)
                    line_dict[key] = self.data.iloc[t:length]
                i += 1

            # returns the key-value pairs that are greater than the cut off length, so no short lines
            line_dict = {key: val for key, val in line_dict.items() if len(
                line_dict[key]) > cutoff_length}
            
            self.lines = rename_dict(line_dict)
    
    def get_heading(self):
        cols = self.data.columns.values
        if 'Easting' in cols and 'Northing' in cols:
            compass = []
            for i in range(len(self.data.index)-1):
                pointa = (
                    self.data.Easting.values[i],self.data.Northing.values[i])
                pointb = (
                    self.data.Easting.values[i+1],self.data.Northing.values[i+1])
                compass.append(
                    direction_lookup(
                        pointb[0], pointa[0], pointb[1], pointa[1])
                )
            compass.insert(0, 999)
        else:
            print("Data must be transormed to UTM")
            exit()

        self.data["Heading"] = compass
            
    def cut_NS(self,buffer = 5):

        length = len(self.data.index)
        cond = []

        for i in range(length):
            if self.data.Heading[i] < 0+buffer or self.data.Heading[i] > 360-buffer\
                    or self.data.Heading[i] > 180-buffer and self.data.Heading[i] < 180+buffer:

                cond.append(True)
            else:
                cond.append(False)

        self.data = self.data[cond].reset_index(drop=True)
    
    def cut_EW(self, buffer: int = 5):

        length = len(self.data.index)
        cond = []

        for i in range(length):
            if self.data.Heading[i] > 270-buffer and self.data.Heading[i] < 270+buffer\
                or self.data.Heading[i] > 90-buffer and self.data.Heading[i] < 90+buffer:

                cond.append(True)
            else:
                cond.append(False)

        self.data = self.data[cond].reset_index(drop=True)
        
    def subtract_total_field(self,value):
        self.data.Mag_nT -= value

    def subtract_mean(self):
        self.data.Mag_nT -=  self.data['Mag_nT'].mean() 

    def get_single_line(self, line_params,line_polygon):
        
        dfp = GeoDataFrame(self.data,geometry= self.data.loc[:,["Easting","Northing"]].apply(Point, axis=1))
        
        line_dict = {}
        for i,line in enumerate(line_polygon):
            name = 'line '+str(i+1)

            actual_line = GeoDataFrame(geometry=line).sjoin(dfp)
            
            line_dict[name] =  actual_line.iloc[:,2:].reset_index(drop=True)
            
        self.lines = line_dict    
        
    def detrend(self,key_name):

        self.lines[key_name].sort_values(by=['Northing'], ascending=True,inplace=True)
        self.lines[key_name].reset_index(drop=True,inplace=True)
        
        correction_line = make_detrend_line(self.lines[key_name])
        
        detrend_mag =[]
        for i,_ in enumerate(self.lines[key_name].index):
            detrend_mag.append(
                self.lines[key_name].Mag_nT.values[i]-correction_line[i]
            )

        self.lines[key_name]['Mag_nT'] = detrend_mag
        
```

## Helper Functions

These functions add some additional fuinctionality that is not within the scope of the previous python class

```python
def files_to_dict(dirpath):
    num = 0
    files_dict = {}
    for filename in listdir(dirpath):
        if '.' in filename: 
            num += 1
            fullpath = Path(dirpath+filename)
            files_dict[str(num)] = fullpath

    return files_dict

def merge_object_data(dictionary):
    data_list =[]
    for key,value in dictionary.items():
        data_list.append(dictionary[key])

    return pd.concat(data_list)

def direction_lookup(destination_x: float, origin_x: float,
                      destination_y: float, origin_y: float) -> float:
    # CREDIT: https://www.analytics-link.com/post/2018/08/21/calculating-the-compass-direction-between-two-points-in-python
    deltaX = destination_x - origin_x

    deltaY = destination_y - origin_y

    degrees_temp = atan2(deltaX, deltaY)/pi*180

    if degrees_temp < 0:

        degrees_final = 360 + degrees_temp

    else:

        degrees_final = degrees_temp

    return degrees_final

def rename_dict(dictionary):
    i = 0
    new_dict={}
    for key,_ in dictionary.items():
        i+=1
        if i > 9:
            newkey = 'line_{}'.format(i)
        else:
            newkey = 'line {}'.format(i)
           
        if newkey!=key:  
            new_dict[newkey] = dictionary[key]
        else:
            new_dict[key] = dictionary[key]
    return new_dict

def line_to_polygon(line_params,buffer=10):
        lines=[]
        for i, (key,value) in enumerate(line_params.items()):
            start = line_params[key][0]
            end = line_params[key][1]

            line = LineString([start,end])
            line = (GeoSeries([line]).buffer(buffer)
)
            lines.append(line)

        return lines

def make_detrend_line(data):
    X1, Y1 = data.Northing.min(), data.Mag_nT.iloc[0]
    X2, Y2 = data.Northing.max(), data.Mag_nT.iloc[-1]
    dy = (Y2-Y1)
    dx = (X2-X1)
    m = (dy/dx)
    b = Y1 - (m*X1)

    correction_line = []
    for i,_ in enumerate(data.index):
        mx = data.Northing.values[i] *m
        y = mx +b 
        correction_line.append(y)

    return correction_line

```

## Plotting Functions

This cell holds all of the plotting functions I would like to use for this data and takes as the primary argument a MagApp with some optional variables based on the specific instance. Functions are matplotlib based.

```python
# Simple plot is wirtten to accept the MagApp class and 
# do the plotting for you based on what the attributes 
# in the object looks like

def simple_plot(application,lines = False):

    if not application.parameters:
        path = 'Merged Files'
    else:
        path = application.parameters['filepath']
        path = str(path).split('/')[-1]
    
    fig, ax = plt.subplots(figsize=(10, 10))

    if not lines:
        data = application.data 
        cols = data.columns
        if 'Easting' in cols:
            ax.plot(data.Easting, data.Northing,
                    linestyle='None', marker="o", ms=2)
            ax.set_title(path+': all data')
        else:
            ax.plot(data.Long, data.Lat,
                    linestyle='None', marker="o", ms=2)
            ax.set_title(path+': all data')
        plt.show()

    else:
        data = application.lines
        for key in data:
            ax.plot(data[key].Easting,data[key].Northing,
                    linestyle='None', marker="o", ms=2,label=key)
            
            ax.set_title(path+': all lines')
            plt.legend(title = "Key/Line Names:",loc='lower right')
        plt.show()

# plot_mag_profile is wirtten to accept the MagApp class, and what line you would like to plot, 
# the line direction 'NS' or 'EW' also needs to be specified, 
def plot_mag_profile(application,key_name,direction) -> None:

    path = application.parameters['filepath']
    path = str(path).split('/')[-1]

    data = application.lines[key_name]

    fig, ax = plt.subplots(figsize=(15, 5))

    ax.set_title(path +': '+key_name)

    if direction != 'NS':
        ax.plot(data.Easting, data.Mag_nT, marker="o",
                linestyle='None', markersize=3)

        ax.set_xlabel("Easting")
        ax.set_ylabel("Magnetic Signal (nT)")

        plt.show()

    else:
        ax.plot(data.Northing, data.Mag_nT, marker="o",
                linestyle='None', markersize=3)

        ax.set_xlabel("Northing")
        ax.set_ylabel("Magnetic Signal (nT)")

        plt.show()

# plot_offset_profile is wirtten to accept the MagApp class, the line direction 'NS' or 'EW'
# needs to be specified, the offset between lines has a default value so each line can be seen 
def plot_offset_profile(application,direction,offset=150) -> None:

    data = application.lines

    path = application.parameters['filepath']
    path = str(path).split('/')[-1]

    start_offset = 0

    fig, ax = plt.subplots(figsize=(10, 10))
    

    if direction != 'NS':
        for key in data:
            ax.plot(data[key].Easting, data[key].Mag_nT+start_offset, marker="o", linestyle='None',
                    markersize=3, label=str(key))

            ax.set_xlabel("Northing")
            ax.set_ylabel("Magnetic Signal (nT)")

            ax.set_title(path+' - Lines offset by: '+str(offset)+'nT')

            plt.legend(title = "Key/Line Names:",loc='lower right')
            start_offset += offset
    else:
        for key in data:
            ax.plot(data[key].Northing, data[key].Mag_nT+start_offset, marker="o", linestyle='None',
                    markersize=3, label=str(key))

            ax.set_xlabel("Northing")
            ax.set_ylabel("Magnetic Signal (nT)")

            ax.set_title(path+' - Lines offset by: '+str(offset)+'nT')

            plt.legend(title = "Key/Line Names:",loc='lower right')
            start_offset += offset
    plt.show()
```

## Forward modeling

Based on **Plouff, D., 1976. Gravity and magnetic fields of polygonal prisms and application to magnetic terrain corrections. Geophysics, 41(4), pp.727-741**

### Model parameters

These are the parameters needed to create a PloufModel class

    -LINE: The Dataframe of the line to be modeled 
    -SHAPES: The Dataframe of the shape to use for modeling
    -TOPS: the top bounds in meters of the shape
    -BOTTOM: the bottom bound in meters of the shape
    -INC: the paleoinclination from the surveyed location
    -DEC: the paleodeclination from the surveyed location
    -INT: the magnetic intensity for the model


```python
class PloufModel():

    def __init__(self,line,shape,top_bound,bottom_bound,inclination,declination,intensity):
        
        self.line = line 
        self.shape = shape
        self.top_bound = top_bound
        self.bottom_bound = bottom_bound
        self.inclination = inclination
        self.declination = declination
        self.intensity = intensity
        self.results = None
        self.residuals = None

    def center_line(self,x0: int, y0: int):

        #Centering data around the middle of Survey Line  
        self.line['Northing'] -= x0
        self.line['Easting'] -= y0

        self.line = self.line


    def center_shapes(self,x0: int, y0: int):

        self.shape['Northing'] -= x0
        self.shape['Easting'] -= y0

        self.shape = self.shape

    def plouf(self,X,Y,z1):
 
        #Setting Parameters
        # print('Running Plouf...')
        x = X
        y = Y
        z2 = self.bottom_bound
        sides=len(X)-1
        df_model = pd.DataFrame()

        #x must be the northing (north coordinate), 
        #y must be the easting (east coordinate),
        #Map coordinates, must be clockwise 

        minc = self.inclination*(pi/180) #down in rad 
        mdec = self.declination*(pi/180) #east in rad

        #calculate direction cosines of magnetiation
        ml = cos(minc) * cos(mdec)
        mm = cos(minc) * sin(mdec)
        mn = sin(minc)

        #components of magnetization in x,y,z, directions

        mx = self.intensity*ml
        my = self.intensity*mm
        mz = self.intensity*mn

        # set earths field
        einc = 62*(pi/180)
        edec = 12*(pi/180)

        el = cos(einc) * cos(edec)
        em = cos(einc) * sin(edec)
        en = sin(einc)

        #proportionality constant 

        prop = 400*pi

        northing_list =[]
        easting_list =[]
        b_total_list =[]
    ########################################################
        #insert for J loop
        for j,k in zip(self.line.Northing,self.line.Easting):
            #px is the northing of the observation point
            #py is the easting of the observation point
            #northing and easting observation points 
            px = j
            py = k
            
            #setting volume integral to zero 
            v1 = 0
            v2 = 0
            v3 = 0 
            v4 = 0 
            v5 = 0 
            v6 = 0 
            
            for i in range(sides):
                if i == (sides-1):
                    x1 = x[i]-px
                    y1 = y[i]-py
                    x2 = x[0]-px
                    y2 = y[0]-py
                else:
                    x1 = x[i]-px
                    y1 = y[i]-py
                    x2 = x[i+1]-px
                    y2 = y[i+1]-py
                    
                #calculate geometry 
                delta_x = x2-x1
                delta_y = y2-y1
                delta_s = sqrt(delta_x**2 + delta_y**2)
            
                #avoid division by zero if delta s = 0
                if delta_s == 0:
                    delta_s = 0.1
            
                c = delta_y/delta_s
                s = delta_x/delta_s
                p = ((x1*y2) - (x2*y1))/delta_s
                
                d1 = x1*s + y1*c
                d2 = x2*s + y2*c

                r1sq = x1**2 + y1**2
                r2sq = x2**2 + y2**2

                r11 = sqrt(r1sq + z1**2)
                r12 = sqrt(r1sq + z2**2)
                r21 = sqrt(r2sq + z1**2)
                r22 = sqrt(r2sq + z2**2)
                
                f = log((r22+z2)/(r12+z2)*(r11+z1)/(r21+z1))
                q = log((r22+d2)/(r12+d1)*(r11+d1)/(r21+d2))
                w = atan2((z2*d2),(p*r22)) - \
                    atan2((z2*d1),(p*r12)) - \
                    atan2((z1*d2),(p*r21)) + \
                    atan2((z1*d1),(p*r11))
                
                v1 += s*c*f - c*c*w
                v2 += s*c*w + c*c*f
                v3 += c*q
                v4 += -(s*c*f + s*s*w)
                v5 += -(s*q)
                v6 += w

            #calculate the components of the magnetic field 
            bx = prop*(mx*v1 + my*v2 + mz* v3);
            by = prop*(mx*v2 + my*v4 + mz* v5);
            bz = prop*(mx*v3 + my*v5 + mz* v6);

            #calculate total anomaly
            #this calculation works for anomaly << total field strength
            b_total = el*bx + em*by + en*bz
            
            northing_list.append(px)       
            easting_list.append(py)         
            b_total_list.append(b_total)

        df_model['mag'] = b_total_list
        df_model['dist'] = northing_list

        return df_model


    def run_plouf(self):

        x0 = self.line.Northing.median() 
        y0 = self.line.Easting.median()

        self.center_line(x0,y0)
        self.center_shapes(x0,y0)

        shapex = self.shape.Northing.tolist()

        shapey = self.shape.Easting.tolist()

        self.results= self.plouf(shapex,shapey,self.top_bound)

        self.residuals = self.results.copy()

        self.residuals["mag"] = (self.line.Mag_nT - self.results.mag)
```
