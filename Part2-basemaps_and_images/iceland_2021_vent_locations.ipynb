```python
import cartopy.crs as ccrs
from cartopy.io.img_tiles import GoogleTiles
class ShadedReliefESRI(GoogleTiles):
    # shaded relief
    def _image_url(self, tile):
        x, y, z = tile
        url = ('https://server.arcgisonline.com/ArcGIS/rest/services/' \
               'World_Shaded_Relief/MapServer/tile/{z}/{y}/{x}.jpg').format(
               z=z, y=y, x=x)
        return url
```

```python
import matplotlib.pyplot as plt
ax = plt.axes(projection=ShadedReliefESRI().crs)
ax.set_extent([-24.5, -13.5, 63.25, 66.7])
ax.add_image(ShadedReliefESRI(), 8)
ax.plot(-22.2725, 63.905,  marker ="o", markersize = 5, color="red", transform=ccrs.Geodetic())
gl = ax.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
```

```python
from cartopy.io.img_tiles import GoogleTiles
class Imagery(GoogleTiles):
    # shaded relief
    def _image_url(self, tile):
        x, y, z = tile
        url = ('https://server.arcgisonline.com/ArcGIS/rest/services/' \
               'World_Imagery/MapServer/tile/{z}/{y}/{x}.jpg').format(
               z=z, y=y, x=x)
        return url
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection=Imagery().crs)
ax.set_extent([-22.4, -22.1, 63.825, 63.95])
ax.add_image(Imagery(), 12)
ax.plot(-22.2725, 63.905,  marker ="o", markersize = 5, color="red", transform=ccrs.Geodetic())
gl = ax.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
```

```python
import pyproj
isn2016=pyproj.CRS(" +proj=lcc +lat_1=64.25 +lat_2=65.75 +lat_0=65 +lon_0=-19 +x_0=2700000 +y_0=300000")
WGS84=pyproj.CRS("EPSG:4326") # LatLon with WGS84 datum 

vent_locations = [[2539384.43, 180319.79], 
                  [2539363.42, 180280.80], 
                  [2539782.65, 181027.83],
                  [2539750.08, 180951.48],
                  [2539716.68, 180884.78],
                  [2539709.13, 180865.30]]

iceland_grid_east, iceland_grid_north = zip(*vent_locations)
vent_lat, vent_long = pyproj.transform(isn2016, WGS84, iceland_grid_east, iceland_grid_north)

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection=Imagery().crs)
ax.set_extent([-22.3, -22.25, 63.88, 63.9])
ax.add_image(Imagery(), 17)
ax.plot(vent_long, vent_lat,  marker ="o", markersize = 2, color="white", linestyle="", transform=ccrs.Geodetic())
gl = ax.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False

plt.savefig("./iceland_vent_zoom.png", dpi = 300)
plt.show()
```
