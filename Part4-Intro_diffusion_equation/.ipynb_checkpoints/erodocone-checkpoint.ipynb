### A numerical model of scoria cone eruption using the diffusion equation

All models of scoria cone erosion and morphometric dating start with the assumption that the cone has a uniform diffusivity. That is, the cone has a uniform composition (grainsize - degree of non-agglutination (brittle cinders), etc). We want to abandon this assumption by using a diffusivity that varies as a function of depth and/or laterally within the cone, informed by our understanding of the eruptive processes that built the cone. Then we can compare the current topographic profiles, to geologic mapping on the cone, and with the Ar/Ar age determinations.

Personally, I do not think geomorphologic dating of scoria cones works at all! But, if it does, it should have dependency on the types of eruptions that built the scoria cone (from low ballistic transport and lots of agglutination to high columns and deposition of brittle scoria fragments).

In the following, symmetric models are developed that have LAYERS. These layers either have uniform diffusivity (like all previous scoria cone erosion models) or have varying diffusivity (which is much more realistic). Note how the topographic profile changes with time as a function of assumptions about the diffusivity. Note how the bedding dip changes with respect to the topography. It might be that the bedding dip with respect to the topography is a MUCH better indicator of cone age (erosion) than the topographic slope alone. 

### Math

Two-dimensional models of scoria cones are created in cylindrical coordinates. An angle of repose of 30$^{\circ}$ is assumed for the simulated scoria cone, with crater width proportional to height and proportional to diameter of the scoria cone base. In these models the cone is perfectly symmetrical with a closed (unbreached) crater and uniform crater rim height initially, prior to any erosion. We can use any initial topography we want, especially if we go to a 3d model.

The change in shape of the scoria cone with time is governed by erosion. Erosion is usually modeled with the diffusion equation (Arrowsmith, etc) , and these models have been applied to scoria cone erosion (Hooper, etc). In cylindrical coordinates:

$$ \frac{\partial h}{\partial t} = \frac{\alpha}{r} \frac{\partial}{\partial r} \left [ r \frac{\partial h}{\partial r} \right ] $$
where $h$ is the topographic height (m), which varies with radius, $r$ (m). The diffusion coefficient, $\alpha$ (m$^2$\,yr$^{-1}$), governs the rate of cone erosion as a function of time, $t$ (yr).

The idea behind cone erosion models is that morphology (e.g., cone slope) is a function of the age of the cone. This assumes a uniform diffusion coefficient. That is, all cones have the same diffusion coefficient and diffusion coefficients do not change with time. Since the diffusion coefficient is poorly known, there is great uncertainty in the absolute age of a given scoria cone, given its topographic profile. But, relatively, cones in the same climate and with the same diffusion coefficient (brittle scoria) should erode at the same rate. 

Of course, scoria cones are not uniform, but have layers of spatter and brittle scoria, representing a wide range of diffusion coefficients as a function of facies within the scoria cone. 

So, in order for cone morphology to be useful at all as a relative age dating tool, it is necessary to calibrate against the geology of the cone and account for presence of lava lakes, spatter, dikes and variability in welding of scoria, all of which have primary control on the value of the diffusion coefficient.

### Implementing the Math in Python
The differential equation is apprximated using finite differences. First consider the units of the differential equation:

$$ \frac{\partial h}{\partial t} = \frac{\alpha}{r} \frac{\partial}{\partial r} \left [ r \frac{\partial h}{\partial r} \right ] $$

$$ \frac{m}{a} = \frac{m^2/a}{m} \frac{1}{m} \left [ m \frac{m}{m} \right ] $$
where $a$ stands for annum. So the units check out. Now consdier how to approximate trems in the equation:

Approximate the lefthand side of the equation
$$ \frac{\partial h}{\partial t} \approx \frac{h^{t+1}_m - h^{t}_m}{\Delta t} $$

Approximately the righthand side of the equation
$$r_{i,1/2} = \frac{r + \Delta r}{2} $$
$$r_{i,-1/2} = \frac{r - \Delta r}{2} $$

$$h_1 = r_{i,1/2} \times \frac{h^{t}_{i+1} - h^{t}_{i}}{\Delta r}  $$
$$h_2 = r_{i,-1/2} \times \frac{h^{t}_{i} - h^{t}_{i-1}}{\Delta r}  $$

$$\frac{\partial}{\partial r} \left [ r \frac{\partial h}{\partial r} \right ] \approx h_1 - h_2$$

Solve for the future height
$$h^{t+1}_i = \frac{\Delta t \alpha}{r_i} (h_1 - h_2) + h^{t}_i$$

Go ahead and check the units again to make sure no terms are dropped:
$$m = \frac{s \frac{m}{s}}{m} (m) +m$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# erodocone1.pl
# Calculate the height profile of a scoria cone as it changes with time
# using the advection-diffusion equation with radial coordinates.
# C. Connor
# July, 2017
#

timelength = 2e6 #maximum time in years
time_step = 200.0 #time step (yr)
creep = 0.004 #diffusivity in m2/yr
del_r = 2.0 #radial step in meters
totalpoints = 500 #total points on the profile from center outward
crater_width = 100
slope = 30
```

```python

#set initial conditions
#this builds an ideal scoria cone
init_height = np.arange(0,totalpoints,1)
radius = np.arange(0,totalpoints,1)

crater_step = int(crater_width/del_r)
for i in range (0,crater_step):
    radial_distance = float(i) * del_r
    init_height[i] = 100.0 + radial_distance * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
for i in range (crater_step, totalpoints):
    radial_distance = float(i) * del_r
    init_height[i] = init_height[crater_step-1] -  (radial_distance - crater_step*del_r) * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
    if (init_height[i] < 0.0):
        init_height[i] = 0.0

plt.figure(figsize=(10,4))
plt.plot(radius,init_height)
#plt.xlim(0,750)
plt.ylim(-10, 200)
plt.show()
```

```python

height = np.array(init_height, copy=True).astype(np.float32)
new_height = np.arange(0,totalpoints,1).astype(np.float32)

for time in range (1,10000000):
    totaltime = time*time_step;
    if (totaltime > timelength):
        break

    for i in range(1, totalpoints-1):
        radial_distance = float(i) * del_r
        riplusonehalf = (radial_distance + float(i+1)*del_r)/2.0
        riminusonehalf = (radial_distance + float(i-1)*del_r)/2.0
        
        term1 = riplusonehalf *(height[i+1] - height[i])/del_r
        term2 = riminusonehalf *(height[i] - height[i-1])/del_r
        new_height[i] = creep*time_step/radial_distance *(term1-term2) + height[i]
        
    for i in range(1, totalpoints-1):
        height[i] = new_height[i]
        
    height[0] = height[1]
    height[totalpoints-1] = height[totalpoints-2] 
  
```

```python
plt.figure(figsize=(10,4))
plt.plot(radius,height)
plt.plot(radius,init_height)
plt.ylim(-10, 200)
plt.xlabel("radial distance (m)")
plt.ylabel("scoria cone height (m)")
plt.show()
```

### Improvements

1. find a better time step
$$\Delta t = \frac{\Delta r^2}{2 \alpha} $$
2. add layering to better visualize the change in outcrop pattern with time
3. change the diffusivity (creep) of different layers

```python
#find a better time step
timestep = del_r**2/(2*creep)
print ("better time step (yr) = ",timestep)
```

```python
# build a layered cone

#layer 1 ht
init_height_layer1 = np.arange(0,totalpoints,1)
radius = np.arange(0,totalpoints,1)

crater_step = int(crater_width/del_r)
for i in range (0,crater_step):
    radial_distance = float(i) * del_r
    init_height_layer1[i] = 100.0 + radial_distance * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
for i in range (crater_step, totalpoints):
    radial_distance = float(i) * del_r
    init_height_layer1[i] = init_height_layer1[crater_step-1] -  (radial_distance - crater_step*del_r) * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
    if (init_height_layer1[i] < 0.0):
        init_height_layer1[i] = 0.0

#layer 2 ht
init_height_layer2 = np.arange(0,totalpoints,1)
radius = np.arange(0,totalpoints,1)

crater_step = int(crater_width/del_r)
for i in range (0,crater_step):
    radial_distance = float(i) * del_r
    init_height_layer2[i] = 75.0 + radial_distance * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
for i in range (crater_step, totalpoints):
    radial_distance = float(i) * del_r
    init_height_layer2[i] = init_height_layer2[crater_step-1] -  (radial_distance - crater_step*del_r) * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
    if (init_height_layer2[i] < 0.0):
        init_height_layer2[i] = 0.0
        
#layer 3 ht
init_height_layer3 = np.arange(0,totalpoints,1)
radius = np.arange(0,totalpoints,1)

crater_step = int(crater_width/del_r)
for i in range (0,crater_step):
    radial_distance = float(i) * del_r
    init_height_layer3[i] = 50.0 + radial_distance * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
for i in range (crater_step, totalpoints):
    radial_distance = float(i) * del_r
    init_height_layer3[i] = init_height_layer3[crater_step-1] -  (radial_distance - crater_step*del_r) * np.sin(slope*np.pi/180.0)
    radius[i] = radial_distance
    if (init_height_layer3[i] < 0.0):
        init_height_layer3[i] = 0.0
        
#plot it all
plt.figure(figsize=(10,4))
plt.plot(radius,init_height_layer1, "k-")
plt.plot(radius,init_height_layer2, "k-")
plt.plot(radius,init_height_layer3, "k-")

plt.fill_between(radius, init_height_layer1, init_height_layer2, color='grey', alpha=0.5)
plt.fill_between(radius, init_height_layer2,init_height_layer3, color='red', alpha=0.5)
plt.fill_between(radius, init_height_layer3,0, color='yellow', alpha=0.5)

plt.plot(radius,height, "k--")
#plt.xlim(0,750)
plt.ylim(-10, 200)
plt.xlabel("radial distance (m)")
plt.ylabel("scoria cone height (m)")
plt.show()
```

A logical follow-on question is to consider variations in diffusivity (creep) of layers within the cone. What if some layers are more difficult to erode than others. For example, some layers might be near vent agglutinate and others loose, unconsolidated tephra. These differences in layers must effect how the cone erodes with time.

In the following script, diffusivity is allowed to change from layer to layer as a function of depth.

```python
spatter_creep = 0.0004
brittle_creep = 0.004

height_layered = np.array(init_height, copy=True).astype(np.float32)
new_height_layered = np.arange(0,totalpoints,1).astype(np.float32)

for time in range (1,10000000):
    totaltime = time*time_step;
    if (totaltime > timelength):
        break

    for i in range(1, totalpoints-1):
        radial_distance = float(i) * del_r
        riplusonehalf = (radial_distance + float(i+1)*del_r)/2.0
        riminusonehalf = (radial_distance + float(i-1)*del_r)/2.0
        
        term1 = riplusonehalf *(height_layered[i+1] - height_layered[i])/del_r
        term2 = riminusonehalf *(height_layered[i] - height_layered[i-1])/del_r
        
        # condition the diffusivity (creep) on depth
        # account for layers
        if (height_layered[i]< init_height_layer2[i]):
            creep = spatter_creep
        else:
            creep = brittle_creep
            
        new_height_layered[i] = creep*time_step/radial_distance *(term1-term2) + height_layered[i]
        
    for i in range(1, totalpoints-1):
        height_layered[i] = new_height_layered[i]
        
    height_layered[0] = height_layered[1]
    height_layered[totalpoints-1] = height_layered[totalpoints-2] 

```

```python
#plot it all
plt.figure(figsize=(10,4))
plt.plot(radius,init_height_layer1, "k-")
plt.plot(radius,init_height_layer2, "k-")
plt.plot(radius,init_height_layer3, "k-")

plt.fill_between(radius, init_height_layer1, init_height_layer2, color='grey', alpha=0.5)
plt.fill_between(radius, init_height_layer2,init_height_layer3, color='red', alpha=0.5)
plt.fill_between(radius, init_height_layer3,0, color='yellow', alpha=0.5)

plt.plot(radius,height_layered, "k--")
plt.plot(radius, height, "r--")
#plt.xlim(0,750)
plt.ylim(-10, 200)
plt.xlabel("radial distance (m)")
plt.ylabel("scoria cone height (m)")
plt.show()
```
