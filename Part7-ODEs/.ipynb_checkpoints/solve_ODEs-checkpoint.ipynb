### Numerical solutions to ODEs

Start with a simple ordinary differential equation:

$$\frac{dy}{dx} = y $$

with initial condition of y = 1 at x = 0

In this case there is an exact solution that is readily obtained:
$$ \int \frac{dy}{y} = \int dx $$

$$ \ln y = x $$

$$y = \exp(x) $$

We will use the exact solution to verify the application of numerical approximations.

Say, $x$ is time and $y$ is number of rabbits. The rate of change in number of rabbits is equal to the current number of rabbits. As the number of rabbits increases, the rate of growth increases. This is exponential growth.

You never know where rabbits will crop up on volcanoes:



image by Kozyndan, 2010.

Notice that this ODE naturally gives rise to the equation of a hyperbola ($a = 1/b$), and the integral (the area under the hyperbola) scales by the factor $e=2.71828$. Euler discovered this relationship. Bernoulli previously identified the value of the approximate value of the transcendental number $e$ while studying limits of infinite series.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,.1)

plt.plot(x,np.exp(x))
plt.xlabel("time")
plt.ylabel("number of rabbits")
plt.show()
```

The following is devoted to development of methods to approximate the solution of the ODE using numerical rather than analytical methods. This helps especially when the solution to an ODE is difficult or unknown. 



### Euler's method of approximating the solution to an ODE
The "original" method for solving ODEs numerically was Euler's method. It is not used because of approximation errors, but it is useful to illustrate more practical methods that developed starting with Euler's approach.

Given an initial value of $y$, what is the value of $y$ after $x$ time or distance is covered? Start with known values and extrapolate forward in small increments. That is:

$$y_{i+1} = y{i} + h f(x,y)$$

Note the error increases as the step size, $h$, increases. Also note the total error increases as the total number of increments increases. 

```python
# implement Euler's method in code 
# dy/dx = y, y = 1 at x = 0

import numpy as np
def f(x,y): 
    return y
x,y = 0,1
h = 0.01
while x<10.:
    k1 = f(x,y)
    y += h*k1
    x += h

y_exact = np.exp(x)
print('Euler method: %.16f, exact: %.16f, diff: %.16f' % (y,y_exact,abs(y-y_exact)))
```

### Runge - Kutta
Euler's method leads to extrapolation errors. To reduce these errors, use the Runge-Kutta method.

The 2nd order Runge-Kutta method is just like Euler's method but cleverly calls the function twice, approximating on the 1/2 step (1/2 h).

$$k_1 = f(x,y) $$
$$k_2 = f(x_{1/2},k_1  y_{1/2}) $$
$$y_{i+1} = y_{i} + h k_2$$

```python
# implement the second order Runge-Kutta solution to
# dy/dx = y. initial condition y = 1 at x = 0
# use a small step size (h = 0.001)... x, x+0.001, ...
x,y = 0,1
h = 0.001
while x<10.:
    k1 = f(x,y)
    k2 = f(x+0.5*h, y+0.5*h*k1)
    y += h*k2
    x += h

y_exact = np.exp(x)
print('RK2 method: %.16f, exact: %.16f, diff: %.16f' % \
(y,y_exact,abs(y-y_exact)))
```

From second order Runge-Kutta, simply add approximation steps to achieve higher accuracy with higher order Runge-Kutta methods. The fourth order method is generally great for obtaining a fast solution that is accurate if the problem is not too large.

Fourth order Runge-Kutta:

```python
# implement the fourth order Runge-Kutta solution to
# dy/dx = y. initial condition y = 1 at x = 0
# use a small step size (h = 0.001)... x, x+0.001, ...

x,y = 0,1
h = 0.001
while x<10.:
    k1 = f(x,y)
    k2 = f(x+0.5*h, y+0.5*h*k1)
    k3 = f(x+0.5*h, y+0.5*h*k2)
    k4 = f(x+h, y+h*k3)
    y += h/6.*(k1+2.*k2+2.*k3+k4)
    x += h

y_exact = np.exp(x)
print('RK4 method: %.16f, exact: %.16f, diff: %.16f' % \
(y,y_exact,abs(y-y_exact)))
```

The fifth order Runge Kutta with refinement of the stepsize is the method you probably want to implement in some form for your specific problem if the problem is large (e.g., required many small steps).

The constants used for $k1 - k6$ are found using the Cash-Karp method (see Numerical Receipes in C) for explanation of these constants.

```python
x, y = 0., 1.
h = 0.001
steps = 0
ct = 0
while x<10.:
    while(True):
        k1 = f(x,y)
        k2 = f(x+1./5.*h, y+1./5.*h*k1)
        k3 = f(x+3./10.*h, y+h*(3./40.*k1 + 9./40.*k2))
        k4 = f(x+3./5.*h, y+h*(3./10.*k1 - 9./10.*k2 + 6./5.*k3))
        k5 = f(x+h,y+h*(-11./54.*k1 + 5./2.*k2 - 70./27.*k3 + 35./27.*k4))
        k6 = f(x+7./8.*h,y+h*(1631./55296.*k1 + 175./512.*k2 + 575./13824.*k3 + 44275./110592.*k4 + 253./4096.*k5))
        yn = y+h*(37./378.*k1 + 250./621.*k3 + 125./594.*k4 + 512./1771.*k6)
        yp = y+h*(2825./27648.*k1 + 18575./48384.*k3 + 13525./55296.*k4 + 277./14336.*k5 + 1./4.*k6) 
        ct+=1
        err = max(abs(yn-yp)/1E-14,0.01)
        if err<1.: break

        hn = 0.9*h*err**-0.25
        if hn < h*0.1: hn = h*0.1
        h = hn 
        
    y = yn
    x += h
    steps += 1
    hn = 0.9*h*err**-0.2
    if hn > h*5.: hn = h*5.
    h = hn 
print("count =", ct)
y_exact = np.exp(x)
print('RK45 method after %d step (t=%.16f): %.16f, exact: %.16f, diff: %.16f' % (steps,x,y,y_exact,abs(y-y_exact)))
```

#### Problem

Consider a magma reservoir with volume 0.1 km$^3$ containing a basaltic magma of density 2800 kg/m$^3$. Suppose the volcano is erupting magma at a steady rate of 5 m$^3$/s. In turn, primitive magma is being supplied to the magma reservoir at 5m$^3$/s, so the volume of the magma reservoir does not change with time. Given the primitive magma has MgO concentration of 16 wt%, and the original magma reservoir composition was 8 wt% MgO, what is the change in MgO concentration in the erupted lavas after 1/2 yr eruption? Imagine that the magma reservoir is well-mixed and there are no other mechanisms of MgO gain or loss.

#### Solution 
This is an example of a "tank" problem. Let $x(t)$ be the mass of MgO in the magma reservoir. The rate of change of MgO in the reservoir, $x'(t)$, is equal to the rate at which MgO enters the reservoir minus the rate at which MgO erupts from the reservoir. The rate at which MgO enters the reservoir is (5 m$^3$/s)(2800 kg/m$^3$)(0.16) = 2240 kg/s. The rate at which MgO leaves the reservoir is equal to the eruption rate times the density times the concentration of MgO in the magma. The concentration of MgO is x(t) divided by the total reservoir mass (x(t)/2.8e11 kg). Hence the rate of outflow of MgO is (14000 kg/s)(x(t)/2.8e11 kg). 

The initial condition is:

$$ x(0) = 0.08 \times 2.8 \times 10^{11} = 2.24 \times 10^{10} \textrm{ kg MgO in reservoir}$$

The differential equation is:

$$x'(t) = \frac{dx}{dt} =  2240 \textrm{ kg/s} - (14000 \textrm{ kg/s})\frac{x(t) \textrm{ kg}}{2.8 \times 10^{11} \textrm{ kg}}$$

This equation has an analytical soution. Implementing the solution in code, using R-K45:

```python
# the ODE to solve
def f(t,x):
    return 2240 - 14000/2.8e11 * x
```

```python
def RK45(f, x, y, h, total_x):
    while x<total_x:
        while(True):
            k1 = f(x,y)
            k2 = f(x+1./5.*h, y+1./5.*h*k1)
            k3 = f(x+3./10.*h, y+h*(3./40.*k1 + 9./40.*k2))
            k4 = f(x+3./5.*h, y+h*(3./10.*k1 - 9./10.*k2 + 6./5.*k3))
            k5 = f(x+h,y+h*(-11./54.*k1 + 5./2.*k2 - 70./27.*k3 + 35./27.*k4))
            k6 = f(x+7./8.*h,y+h*(1631./55296.*k1 + 175./512.*k2 + 575./13824.*k3 + 44275./110592.*k4 + 253./4096.*k5))
            yn = y+h*(37./378.*k1 + 250./621.*k3 + 125./594.*k4 + 512./1771.*k6)
            yp = y+h*(2825./27648.*k1 + 18575./48384.*k3 + 13525./55296.*k4 + 277./14336.*k5 + 1./4.*k6) 
        
            err = max(abs(yn-yp)/1E-14,0.01)
            if err<1.: break

            hn = 0.9*h*err**-0.25
            if hn < h*0.1: hn = h*0.1
            h = hn 

        y = yn
        x += h
        
        hn = 0.9*h*err**-0.2
        if hn > h*5.: hn = h*5.
        h = hn 
    return y

#Initial conditions
t = 0  #at zero seconds
x = 2.24e10 # mass of MgO in reservoir at zero seconds
h = 10000 #initial time step (s)

#total simulation time
x_total = 0.5*365*24*3600 #run for 0.5 yr

#solve the ODE
y_total = RK45(f,t, x,h,x_total)
print("MgO concentration after 1/2 yr = ",y_total/2.8e11) #divide by mass of reservoir to get MgO concentration
```

The following implements the same code for the same problem, but builds a list in the RK45 function to show the time history of concentration in the magma reservoir. Then the output time history is plotted in the next cell.

```python
def RK45(f, x, y, h, total_x):
    xout=[] #initialize lists for the output
    yout = []
    while x<total_x:
        while(True):
            k1 = f(x,y)
            k2 = f(x+1./5.*h, y+1./5.*h*k1)
            k3 = f(x+3./10.*h, y+h*(3./40.*k1 + 9./40.*k2))
            k4 = f(x+3./5.*h, y+h*(3./10.*k1 - 9./10.*k2 + 6./5.*k3))
            k5 = f(x+h,y+h*(-11./54.*k1 + 5./2.*k2 - 70./27.*k3 + 35./27.*k4))
            k6 = f(x+7./8.*h,y+h*(1631./55296.*k1 + 175./512.*k2 + 575./13824.*k3 + 44275./110592.*k4 + 253./4096.*k5))
            yn = y+h*(37./378.*k1 + 250./621.*k3 + 125./594.*k4 + 512./1771.*k6)
            yp = y+h*(2825./27648.*k1 + 18575./48384.*k3 + 13525./55296.*k4 + 277./14336.*k5 + 1./4.*k6) 
        
            err = max(abs(yn-yp)/1E-14,0.01)
            if err<1.: break

            hn = 0.9*h*err**-0.25
            if hn < h*0.1: hn = h*0.1
            h = hn 

        y = yn
        x += h
        xout.append(x) # concatenate the lists
        yout.append(y)
        
        hn = 0.9*h*err**-0.2
        if hn > h*5.: hn = h*5.
        h = hn 
    return y, xout, yout

#initial conditions
t = 0  #at zero seconds
x = 2.24e10 # mass of MgO in reservoir at zero seconds
h = 10000 #initial time step (s)

#run time of the simulation
t_total = 5*365*24*3600 #run for 5 yr

# solve the ODE
x_total, tout, xout = RK45(f,t, x,h,t_total)

#scale the output data to managable units
tout = np.array(tout)/(365*24*3600) # normalize output to time in years
xout = np.array(xout)/2.8e11  # normalize output to a concentration
```

```python
#plot the ODE solution output
import matplotlib.pyplot as plt

plt.plot(tout,xout)
plt.xlabel("time (yrs)")
plt.ylabel("MgO concentration in erupted lava")
plt.show()
```

### Additional applications

Particles are smashed by collisions as they move in concentrated flows. For example, particles break and are abraded as they flow up volcano conduits, and particles break apart and abrade as they are transported downslope in lahars or PDCs. One model for the change in particle size with distance of transport is called Sternberg's Law. Written as a differential equation:

$$\frac{dw}{dx} = -\frac{w}{a} $$

where $w$ is the particle diameter (seive diameter), $x$ is the distance transported, and $a$ is a constant that governs how easy it is to break up the particles. Pumice has a different value of $a$ than lithic blocks. The idea of the differential equation is that the rate of change in particle size depends on the particle diameter. Large particles are broken apart quickly, finer particles are reduced in size more slowly - an exponential decay law rather than expoential growth law (hence, the minus sign in the ODE).  Sternberg originally applied his relation to grainize in rivers, which inadvertantly convolves the erosion of particles with the natural sorting of particles along $x$ by settling and deposition.
