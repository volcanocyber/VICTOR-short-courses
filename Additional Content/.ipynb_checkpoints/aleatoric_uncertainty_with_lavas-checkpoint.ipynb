### Using scipy.stats to look into aleatoric uncertainty

This example will be developed during the VICTOR short course on Wednesday, Feb 8. It continues the discussion of probability in hazard modeling.

Here we focus on the aleatoric uncertainty in hazard assessment. We consider a bunch of lava flow volume data from Nick Peters. These data are lava volume measurements Nick tried to compile from a distributed volcanic field in the western USA (The Eastern Snake River Plain). That compilation is incomplete and the actual numbers are likely to change, but we can consider aleatoric uncertainty using the data.

Specifically, make an empirical exceedence probability curve from these data. Fit a model and consider the aleatoric uncertainty.

```python
#all of Nick's volume data for ESRP lavas (km3)

lava_volume = [0.006790804,
0.018273417,
0.018262274,
1.682772361,
6.010234627,
3.635205175,
3.625225399,
11.36899532,
5.593777,
2.344158262,
4.547043639,
14.51715,
5.90422347,
5.178040265,
0.047037626,
0.001182711,
0.040755318,
0.025170866,
0.002810687,
0.176028838,
0.008080733,
0.000561152,
0.099111753,
0.004345368,
0.044848322,
0.006911029,
0.100695214,
0.06884728,
0.204539571,
0.013020931,
0.09508668,
0.10819065,
0.005805203,
0.002229659,
0.002700955,
0.150743848,
1.961828809]
```

```python
#Nick's volume data look like they follow a log10 Uniform distribution

import matplotlib.pyplot as plt
import numpy as np
plt.hist(np.log10(lava_volume), bins=20)
plt.xlabel("log$_{10}$ lava volume (km$^3$)")
```

```python
x = np.sort(np.log10(lava_volume))
y = 1- np.arange(len(x))/float(len(x))
plt.xlabel("log$_{10}$ lava volume (km$^3$)")
plt.plot(x,y)
```

```python
import scipy.stats as st

log10_volume = np.log10(lava_volume)

min_volume = min(log10_volume) - 0.25
max_volume = max(log10_volume) + 0.1

print("min volume =", 10**min_volume)
print("max volume =", 10**max_volume)
```

```python
lava_threshold = 0.01 # 1 km3
log_lava_threshold = np.log10(lava_threshold)

prob = 1 - st.uniform.cdf(log_lava_threshold, loc=min_volume, scale=max_volume - min_volume)
print(prob)
```

```python
model_x = np.arange(-5,2, 0.1)
model = 1 - st.uniform.cdf(model_x, loc=min_volume, scale=max_volume - min_volume)

upper_bound = model + 1.36/np.sqrt(len(log10_volume))
lower_bound = model - 1.36/np.sqrt(len(log10_volume))
                                   
plt.plot(model_x, model)
plt.plot(x,y)
plt.plot(model_x, upper_bound, linestyle = "dashed")
plt.plot(model_x, lower_bound, linestyle = "dashed")
plt.show()
```
